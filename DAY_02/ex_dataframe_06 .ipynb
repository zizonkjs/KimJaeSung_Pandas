### DataFrame에서 행/열 추가
- 행방향 추가
- 열방향 추가

```python
# 모듈로딩
import pandas as pd
```

```python
# 데이터 준비
data={'name':['홍길동', '마징가', '원더우먼'],
       'job':['의적','로봇', '영웅'],
       'gender':['남','남','여'],
       'age':[100, 200, 300],
       'height':[180,300,200]}
```

```python
dataDF=pd.DataFrame(data)
dataDF
```

- 1번행 추가 : 변수명.loc['추가할행이름'] = 데이터**
- iloc는 사용할수없음.
- 데이터수 == columns 경우엔 행추가가능!
- 데이터수 == 1개 경우 행추가 가능!!
- 그 외의 경우, 행추가 X


```python
# 한개행 추가 -> 데이터 수가 열 갯수와 동일한 경우
dataDF.loc[3]=['베트맨', '박쥐','남',40,190]
dataDF.loc[4]=['베트맨', '박쥐','남',40,190]
```

```python
dataDF
```

```python
# 한개행 추가 -> 데이터 수를 한개로
dataDF.loc[5]=100
dataDF # 데이터를 하나만 주면 모든 열에 같은 값이 저장됨.
```

```python
# 1개 행 추가 => 데이터 수를 2개로

# dataDF.loc[6]=['스파이더맨', '거미']
# dataDF 
# 줄거면 갯수를 맞춰서 주던가 or 한개만 주던가 둘중에 하나만 해야함.
```

- 열 추가 : 변수명.loc['추가할컬럼이름'] = 데이터**
- iloc는 사용할수없음.
- 데이터수 == 행 수때엔 컬럼추가가능!
- 데이터수 == 1개 경우 컬럼추가 가능!!
- 그 외의 경우, 컬럼추가 X

```python
# age 컬럼을 추가하기 : 행 개수 == 데이터 개수
# weight 추가
dataDF['weight']=[90, 100, 80, 110, 120, 150]
dataDF
```

```python
# age 컬럼을 추가하기 : 1개 == 데이터 개수
dataDF['weight2']=0
dataDF
```
