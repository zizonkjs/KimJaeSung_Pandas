## [ 결측치 처리 ]
- 빈 칸 즉, 데이터가 없는 부분을 의미 
- 결측치 체크 후 처리 방법 결정 
- 처리 방법  

    -제거  

    -치환  
    

```python
# 1. 모듈로딩
import pandas as pd
```

```python
# 2. 데이터 준비
file_name='data.csv'
```

```python
# 3. csv 파일을 데이터 프레임으로 만들기
datadf=pd.read_csv(file_name)
```

```python
# 4. 데이터 기본 정보 확인
datadf.info()
```

```python
# 실제 데이터와 컬럼별 타입 체크
datadf.head()
```

```python
# 5. 데이터 전처리
# 5-1. 결측치 체크-> isnull(), isna() 메서드
datadf.isna()
```

```python
datadf.isna().sum()
```

```python
# -결측지 처리 -> 삭제
# 기본값 : 행방향, 행 안에 1개의 커럼이라도 Na이면 삭제
datadf.dropna()
```

```python
# 결측치 처리를 열방향으로 정리
datadf.dropna(axis=1)
```

```python
# 설정값 : 모든 원소가 NaN일때 삭제 how_=all
datadf.dropna(axis=0, how='all') # 모든 원소가 NaN인게 없기때문에 안지워짐.
```

```python
# 설정값 : 정상데이터가 지정된 개수 이하이면 삭제 (thresh=정수)
datadf.dropna(axis=0, thresh=4)
```

```python
# 특정 컬럼만 NaN여부 검사 진행 후 결정 : subset
# 나이를 분석하고싶음. -> age 컬럼만 보면됨
datadf.dropna(axis=0, subset=['age'])
```

```python
# 나이와 이름을 분석하고싶음.
# datadf.dropna(subset=['name'], axis=0)
```

```python
# 결측치 처리 -> 방법 2) 대체 치환 fillna() 메서드
# - 대체 값 : 원하는 값 or  컬럼의 평균 값, 최빈값 등을 사용할 수 있음.
#            이전의 값 또는 이후의 값으로 대체
datadf.fillna(0)
```

```python
datadf.fillna({'name':'unknown', 'age':0})
```

```python
# 위에 커럼의 값으로 집어넣기
datadf.ffill()
```

```python
# 아래 컬럼의 값으로 채우기
datadf.bfill()
```

```python
# 나이의 결측치는 나이 평균으로 채우기
# nae=datadf['age'].mean()
# datadf.fillna({'age':14.0})
datadf.fillna({'age':datadf['age'].mean()})
```
