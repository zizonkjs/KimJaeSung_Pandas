### [matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# 모듈 로딩
# 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기 - 선(Line) 그래프
# plot(기본값:y축데이터)
lineobj=plt.plot([1,2,3,4]) # 선그래프
plt.xlabel('DATA')
plt.ylabel('DATA2')
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(X축데이터, Y축데이터)
plt.plot([10,20,30], [5,10,15])
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30], [5,10,15], 'b--')
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30], [5,10,15], 'ro--') # x와 y가 만나는지점에 점
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30], [5,10,15], 'k<--') 
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30], [5,10,15], 'go') 
plt.show()
```

```python
# 그래프 그리기 - Line(선) 그래프
# plot(X축데이터, Y축데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10,20,30], [5,10,15], 'r+') 
plt.show()
```

```python
# linewidth=2, markersize=12 두깨 마크사이즈 넣어주기
# plot(X축데이터, Y축데이터, line2D속성으로 설정)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
# mec, mew 사용
plt.plot([10,20,30], [5,10,15], color='red', marker='o', linestyle='dashed', markersize=10, linewidth=4, mec='blue', mew=5) 
plt.show()
```

```python
# 3. 그래프의 부가적인 기능을 추가
# - 그래프 제목 
# - x축 과 y축 라벨
# - 여러개의 그래프일 경우 범례 -> 그래프의 라벨 속성 값을 범례에서 사용
#                                loc 속성 위치 지정 가능 [기본값:best]
plt.plot([10,20,30],[15,25,20],'bo--', label="V") # label 있으면 legend 안써도됨.
plt.plot([10,20,30],[15,55,25],'ro--', label="2")
plt.title('[MPLIB LINE2D]') # 표 제목
plt.xlabel("Data")
plt.ylabel('Result')
plt.legend() # 범례 legend(['First line', 'Second line'])
plt.show()
```
