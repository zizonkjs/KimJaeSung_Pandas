### [ Matplotlib 활용 시각화 ]

- x축과 y축 다양한 처리


```python
# 모듈 로딩
from matplotlib import pyplot as plt
```

```python
# 그래프 그리기 x,y 축의 눈금(tick) 설정, figure를 자동 맞춤
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']


plt.plot(x_data, y_data, marker='o')
# plt.xticks(rotation=45)
plt.tight_layout() # 알아서 맞춰줌 모든 내용 출력되게 자동 맞춤 설정해줌.
plt.show()
```

```python
# 그래프 그리기 - 눈금(ticks) 설정
y_data = [1,3,5,7,9]
x_data = ['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']


plt.plot(y_data, marker='o')
plt.ylim(bottom=0, top=15)
plt.xlim(left=-1, right=7) # 간격설정

font_config={'fontsize':22, 'fontweight':'bold'}
plt.xticks([0,2,4,6],['A','B','C','D'], **font_config)  # 눈금설정,[눈금 라벨1,2,3,4]
plt.show()
```
