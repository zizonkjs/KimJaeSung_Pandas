### [ 그룹화 ]
- Sereis/DataFrame에 groupby() 메서드를 사용
- Series/DataFrameGroupBy Object 결과 반환

```python
# 모듈 로딩
import pandas as pd
import DAY_07.df_dfutil as ut
```

```python
# 데이터 준비
file_name='titanic.csv.csv'
datadf=pd.read_csv(file_name)
datadf
```

```python
# 데이터 확인
ut.checkDataFrame(datadf, 'datadf')
```

```python
# 분석주제 --> 성별에 따른 나이 분포
# 분석컬럼 즉, 성별과 나이컬럼 필요
# loc[행,열]
gender_agedf=datadf.loc[:, ['성별', '나이']] # datadf[['sex','age']]
```

```python
# 데이터 전처리 : 결측치, 중복값
# 결측치 : isna(), isnull()
gender_agedf.isna().sum()
gender_agedf
```

```python
# 결측치 처리 -> 성별에 따라서 치환하기
# 성별에 따른 그룹화
groupobj=gender_agedf.groupby('성별') # 성별을 기준으로 그룹화해라
```

```python
# Group 객체의 속성
# groups 속성 : 그룹화된 그룹에 속하는 인덱스정보를 저장 Dict형태
print(groupobj.groups.keys())
print(groupobj.groups.values())
```

```python
# indices 속성 : 각 그룹의 index 속성을 저장
print(groupobj.indices)
print(groupobj.indices.keys())
```

```python
# get_group() method : 그룹화된 것들중에서 특정 그룹 데이터 읽기.
for key in groupobj.groups.keys():
    print(groupobj.get_group(key).ndim, groupobj.get_group(key).shape) # 반환값이 DataFrame
    print(groupobj.get_group(key).min(), groupobj.get_group(key).max())
```

```python
# 그룹별로 집계연산
print(groupobj.count())
```

```python
print(groupobj.max())
```

```python
print(groupobj.min())
```

```python
print(groupobj.mean())
```

```python
print(groupobj.median())
```

```python
resultdf=groupobj.agg(['min', 'max' , 'mean', 'median','sum'])
resultdf
```
