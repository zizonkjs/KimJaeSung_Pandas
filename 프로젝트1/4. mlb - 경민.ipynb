MLB 주루 포지션 분석
  * 주요 용어 분석

    - hp_to_1b :  홈 플레이트에서 1루까지 속도

    - sprint_speed : 선수의 스프린트 속도

    - total_stolen_base : 도루 성공 횟수 

    - total_caught_stealing : 도루 실패 횟수 
    
    - ab_scoring : 득점으로 이어진 타석 

    - run : 득점 수


```python
# 모듈 로딩
import pandas as pd 
import numpy as np
```

```python
# 파일 로딩
file = 'baseball.csv'
mlb = pd.read_csv(file)
mlb
```

```python
# mlb 데이터 확인
mlb.head()
```

```python
# 파일 정보 
mlb.info()
```

```python
# 컬럼 이름 변경
# r_total_stolen_base --> total_stolen_base로 바꿔주기 
# r_total_caught_stealing --> total_caught_stealing로 바꿔주기
# b_ab_scoring ---> ab_scoring으로 바꿔주기
# r_run --> run으로 바꿔주기 
mlb.rename({'r_total_stolen_base' : 'total_stolen_base', 'r_total_caught_stealing':'total_caught_stealing','b_ab_scoring':'ab_scoring','r_run':'run'}, axis= 1, inplace=True)
mlb
```

```python
# player_id 제거
mlb.drop(columns='player_id', inplace=True)
mlb
```

```python
# 상관관계 
mlb.corr(numeric_only=True)
```

```python
# 년도별로 데이터 나누기 - 최근 4년 데이터들 (2021 ~ 2024)
mlb_2021 = mlb[mlb['year'] == 2021].reset_index(drop=True).copy()
mlb_2022 = mlb[mlb['year'] == 2022].reset_index(drop=True).copy()
mlb_2023 = mlb[mlb['year'] == 2023].reset_index(drop=True).copy()
mlb_2024 = mlb[mlb['year'] == 2024].reset_index(drop=True).copy()
```

```python
# 2021년 MLB 데이터 확인
display(mlb_2021.corr(numeric_only=True))
display(mlb_2022.corr(numeric_only=True))
display(mlb_2023.corr(numeric_only=True))
mlb_2024.corr(numeric_only=True)
```

```python
mlb_2021_sb = mlb_2021[mlb_2021['total_stolen_base']>=10].reset_index(drop=True)
mlb_2021_sb.sort_values(by=['total_stolen_base'],ascending=False).head()
```

```python
mlb_2022_sb = mlb_2022[mlb_2022['total_stolen_base']>=10].reset_index(drop=True)
mlb_2022_sb.sort_values(by=['total_stolen_base'],ascending=False ,ignore_index=True).head()
```

```python
mlb_2023_sb = mlb_2023[mlb_2023['total_stolen_base']>=10].reset_index(drop=True)
mlb_2023_sb.sort_values(by=['total_stolen_base'],ascending=False,ignore_index=True).head()
```

```python
mlb_2024_sb = mlb_2024[mlb_2024['total_stolen_base']>=10].reset_index(drop=True)
mlb_2024_sb.sort_values(by=['total_stolen_base'],ascending=False,ignore_index=True).head()
```
