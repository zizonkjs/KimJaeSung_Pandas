### [ 파이 그래프 ]
컬럼의 데이터의 분포 비율을 시각화


```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA=r'C:\Users\zizonkjs\pandas\datafolder\auto_mpg.csv'
df=pd.read_csv(F_DATA)
df
```

```python
# 3. 컬럼별 타입 확인
df.dtypes
```

```python
# 4. 특정 컬럼의 데이터 비율
# - 자동차 생산 국가에 대한 비율을 보고싶다.
df['origin'].astype('category')
```

```python
# 자동차 생산국에 대한 비율을 알고싶다 
round(df['origin'].value_counts()/ df['origin'].count(), 2)
orsr=df['origin'].value_counts()
x=orsr.index.tolist()
h=orsr.values.tolist()
x, h
```

```python
# 비율 시각화 - bar 그래프
orgbar=plt.bar(x, h, color=['blue','yellow','green'], edgecolor=['black'])

plt.bar_label(orgbar,h, label_type='center') # 막대위에 숫자넣기, 안에 넣고 싶으면 라벨타입 센터

plt.xticks([1,2,3],['US','EU','JP'], rotation=10, fontsize=20) # ([눈금의 위치값],[설정할이름])
plt.show()
```

```python
# pie 그래프 그려보기 
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%')
plt.show()
```

```python
# pie 그래프 그려보기 
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%', 
        counterclock=False, startangle=270)
plt.show()
```

```python
# pie 그래프 그려보기 
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%', 
        counterclock=False, startangle=90, explode=[0.5,0,0])
plt.show()
```

```python
# pie 그래프 그려보기 
plt.pie(h, labels=['US','EU','JP'], autopct='%.1f%%', 
        counterclock=False, startangle=90, wedgeprops={'linewidth':4, 'width':0.9, 'edgecolor':'w'})
plt.show()
```
